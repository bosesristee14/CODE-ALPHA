```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

# 1. Set up Chrome options to look natural
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled") # Hides the fact that it's an automated script

print("Launching real Chrome browser window...")
driver = webdriver.Chrome(options=options)

# 2. Tell the browser to open the IMDb page
url = "https://www.imdb.com/chart/top/"
driver.get(url)

# 3. Wait 5 seconds to let the page fully load its JavaScript data
print("Waiting for page elements to load...")
time.sleep(5)

# 4. Pass the fully loaded HTML data from Chrome over to BeautifulSoup
html_source = driver.page_source
soup = BeautifulSoup(html_source, "html.parser")

print("Successfully captured the live HTML data!")
import pandas as pd

# 5. Locate the main container blocks for each movie on the chart
# IMDb groups each movie entry inside a specific div component
movie_containers = soup.find_all("div", class_="sc-b189961a-0")

scraped_movies = []

print(f"\nProcessing entries and pulling ratings...")

# 6. Loop through the containers to match titles with ratings
for item in movie_containers[:25]:  # Let's pull the Top 25 for this test!
    try:
        # Extract Title
        title_element = item.find("h3", class_="ipc-title__text")
        # Clean title text (removes numbering prefix like "1. ")
        raw_title = title_element.text.strip().split('. ', 1)[-1] if title_element else "N/A"
        
        # Extract Rating
        rating_element = item.find("span", class_="ipc-rating-star--rating")
        rating = rating_element.text.strip() if rating_element else "N/A"
        
        # Store them together in a dictionary row
        scraped_movies.append({
            "Rank": len(scraped_movies) + 1,
            "Movie Title": raw_title,
            "IMDb Rating": rating
        })
    except Exception as e:
        continue

# 7. Convert our list of rows into a beautiful Pandas DataFrame table
df = pd.DataFrame(scraped_movies)

# 8. Export it to a clean CSV spreadsheet for your manager!
df.to_csv("imdb_top_movies.csv", index=False)

print("\nTask Complete! Here is your custom dataset preview:")
print(df.head(10))
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

# Step 1: Set up Chrome options to hide automation signals
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

print("Launching real Chrome browser window...")
driver = webdriver.Chrome(options=options)

# Step 2: Open the IMDb page
url = "https://www.imdb.com/chart/top/"
driver.get(url)

# Step 3: Wait for the page's JavaScript components to fully display
print("Waiting 5 seconds for page elements to load...")
time.sleep(5)

# Step 4: Transfer the live layout data over to BeautifulSoup
html_source = driver.page_source
soup = BeautifulSoup(html_source, "html.parser")
print("HTML captured successfully! Processing data entries...")

# Step 5: Locate the main card/container blocks for each movie listing
# IMDb groups each listing row inside an 'ipc-metadata-list-summary-item' container
movie_containers = soup.find_all("li", class_="ipc-metadata-list-summary-item")

scraped_movies = []

# Step 6: Loop through the elements to match titles with ratings side-by-side
for item in movie_containers[:25]:  # Gathering the Top 25 entries for this dataset
    try:
        # Extract Title Text
        title_element = item.find("h3", class_="ipc-title__text")
        # Clean title text (removes numbering prefixes like "1. ")
        raw_title = title_element.text.strip().split('. ', 1)[-1] if title_element else "N/A"
        
        # Extract Rating Value
        rating_element = item.find("span", class_="ipc-rating-star--rating")
        rating = rating_element.text.strip() if rating_element else "N/A"
        
        # Append as a structured row dictionary
        scraped_movies.append({
            "Rank": len(scraped_movies) + 1,
            "Movie Title": raw_title,
            "IMDb Rating": rating
        })
    except Exception as e:
        continue

# Step 7: Close the browser window automatically
driver.quit()
print("Browser closed.")

# Step 8: Convert our structured dictionary array into a Pandas DataFrame table
df = pd.DataFrame(scraped_movies)

# Step 9: Export your custom dataset directly to a CSV file
df.to_csv("imdb_top_movies.csv", index=False)

print("\n spreadsheet dataset has been created and saved as 'imdb_top_movies.csv'.")
print("\nHere is a preview of dataset:")
print(df.head(10))
```

```python
# 1. Convert the rating column from text strings to actual decimal numbers (floats)
df["IMDb Rating"] = df["IMDb Rating"].astype(float)

# 2. Calculate the average rating of your custom top list
average_rating = df["IMDb Rating"].mean()

print(f"Dataset Analysis Complete!")
print(f"Total Movies Collected: {len(df)}")
print(f"Average Rating of this Group: {average_rating:.2f}")
```
